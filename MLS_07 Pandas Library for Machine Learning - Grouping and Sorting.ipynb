```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv("data/gun-violence-data1.csv")
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['day_of_week'] = data['date'].dt.dayofweek
data['total_victim'] = data['n_killed']+data['n_injured']
data.head()
```

```python
data['day_of_week'] = data['day_of_week'].map({0:"Monday", 1:"Tue", 2:'Wed', 3:'Thur', 4:"Fri", 5:'Satu', 6:'Sunday'})
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
# grouping
data['day_of_week'].value_counts()
```

```python
data.groupby('day_of_week')['total_victim'].count()
```

```python
data.groupby('state')['total_victim'].sum()
```

```python
import numpy as np
data.groupby('month')['total_victim'].agg([np.mean,min, max])
```

```python
#sorting
data.groupby('state')['total_victim'].sum().sort_values(ascending = False)
```

```python
data.groupby('month')['total_victim'].count().sort_values(ascending = True)
```
